```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="darkgrid")
```

```python
df = pd.read_csv('steam_data.csv')
```

```python
df.columns = ['Titulo', 'Precio original', 'Descuento', 'Fecha de lanzamiento', 'Enlace', 'Descripcion del juego', 'Resumen reseñas recientes', 'Todo el resumen de reseñas', 'Porcentage de reseñas recientes', 'Porcentage global de reseñas', 'Desarollador', 'Publicador', 'Idiomas', 'Tags populares', 'Caracteristicas del juego', 'Requisitos minimos']
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
print(df.isnull().sum())
```

```python
df.drop(columns=['Enlace','Descripcion del juego','Resumen reseñas recientes', 'Porcentage de reseñas recientes', 'Porcentage global de reseñas'])
```

```python
df['Precio original']= df['Precio original'].str.replace('$', '').str.replace(',', '')
df['Precio original'] = df['Precio original'].apply(lambda x: 0 if x == 'Free' else float(x))
```

```python
df['Descuento']= df['Descuento'].str.replace('$', '').str.replace(',', '')
df['Descuento'] = df['Descuento'].apply(lambda x: 0 if x == 'Free' else float(x))
```

```python
df.head()
```

```python
# ¿Si el precio del juego es más elevado, este tendrá reviews más positivas?
# Crear un diagrama de puntos
plt.figure(figsize=(20, 10))
plt.scatter(df['Todo el resumen de reseñas'], df['Precio original'], color='blue', marker='o')

# Añadir etiquetas y título
plt.title('REVIEWS POR PRECIO')
plt.ylabel('Precio original')
plt.xlabel('Reseñas')

# Mostrar el gráfico
plt.grid(True)
plt.show()
```

```python
# ¿Qué compañía es la que cuenta con más entregas a la venta?
# Agrupar por publicador
pubCont = df.groupby('Publicador').size().reset_index(name='Cantidad')
pubCont = pubCont[pubCont['Cantidad'] >= 20]
pubCont.head(3000)

# Crear la gráfica de barras
plt.figure(figsize=(20, 10))
sns.barplot(x='Publicador', y='Cantidad', data=pubCont)

# Añadir títulos y etiquetas
plt.title('CANTIDAD DE JUGOS POR PUBLICADOR')
plt.xlabel('Publicador')
plt.ylabel('Cantidad de Juegos')
plt.xticks(rotation=90)  # Rotar las etiquetas del eje x para mejor visualización

# Mostrar la gráfica
plt.show()
```

```python
# ¿Han salido buenos juegos en 2019?
# Convertir la columna 'Fecha de lanzamiento' al formato de fecha dentro del DataFrame
df['Fecha de lanzamiento'] = pd.to_datetime(df['Fecha de lanzamiento'], format='%d %b, %Y', errors='coerce')

# Filtrar los juegos que tienen fecha de lanzamiento en 2019
juegos2019 = df[df['Fecha de lanzamiento'].dt.year == 2019]

# Agrupar por 'Todo el resumen de reseñas' y contar cuántos juegos hay en cada grupo
juegosAgrupados = juegos2019.groupby('Todo el resumen de reseñas').size().reset_index(name='Cantidad de juegos')

# Crear el gráfico de pastel
plt.figure(figsize=(7,7))
plt.pie(juegosAgrupados['Cantidad de juegos'], labels=juegosAgrupados['Todo el resumen de reseñas'], autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)

# Título del gráfico
plt.title('POPULARIDAD DE LOS LANZAMIENTOS EN 2019')

# Mostrar el gráfico
plt.show()
```

```python
# ¿Qué año ha contado con un major número de juegos en descuento?
# Asegurarse de que 'Fecha de lanzamiento' esté en formato datetime
df['Fecha de lanzamiento'] = pd.to_datetime(df['Fecha de lanzamiento'], format='%d %b, %Y', errors='coerce')

# Filtrar los registros donde 'Precio original' es distinto a 'Descuento'
juegosDesc = df[df['Precio original'] != df['Descuento']].copy()  # Hacemos una copia explícita para evitar el SettingWithCopyWarning

# Extraer el año de la columna 'Fecha de lanzamiento' usando .loc para asignar la columna de forma segura
juegosDesc.loc[:, 'Año de lanzamiento'] = juegosDesc['Fecha de lanzamiento'].dt.year

# Agrupar por 'Año de lanzamiento' y contar el número de registros en cada grupo
juegosDescAgr = juegosDesc.groupby('Año de lanzamiento').size().reset_index(name='Cantidad de juegos')

# Asegurarse de que 'Precio original' y 'Descuento' estén en formato numérico
juegosDesc['Precio original'] = pd.to_numeric(juegosDesc['Precio original'], errors='coerce')
juegosDesc['Descuento'] = pd.to_numeric(juegosDesc['Descuento'], errors='coerce')

# Crear la gráfica de barras
plt.figure(figsize=(20, 10))
plt.barh(juegosDescAgr['Año de lanzamiento'].astype(str), juegosDescAgr['Cantidad de juegos'], color='skyblue')

# Añadir título y etiquetas
plt.title('DESCUENTOS POR AÑO')
plt.ylabel('Año de lanzamiento')
plt.xlabel('Cantidad de juegos')

# Mostrar la gráfica
plt.show()
```

```python
# Convertir la columna 'Fecha de lanzamiento' al formato datetime, manejando errores
df['Fecha de lanzamiento'] = pd.to_datetime(df['Fecha de lanzamiento'], format='%d, %B %Y', errors='coerce')

# Extraer el año de la columna 'Fecha de lanzamiento'
df['Año'] = df['Fecha de lanzamiento'].dt.year

# Eliminar las filas con años nulos o incorrectos
df = df.dropna(subset=['Año'])

# Asegurarse de que la columna 'Año' esté en formato int
df['Año'] = df['Año'].astype(int)

# Agrupar los juegos por año
juegosAño = df.groupby('Año').size().reset_index(name='Cantidad de Juegos')

# Crear gráfico de pastel
plt.figure(figsize=(10, 10))
sizes = juegosAño['Cantidad de Juegos']
labels = juegosAño['Año']

# Generar el gráfico de pastel
plt.pie(sizes, startangle=90, colors=plt.cm.Paired.colors)

# Título del gráfico
plt.title('% ENTREGAS CADA AÑO')

# Crear leyenda con años y porcentajes
# Calcular porcentajes para la leyenda
percentages = (sizes / sizes.sum()) * 100
legend_labels = [f'Año: {year}, %: {percentage:.1f}%' for year, percentage in zip(labels, percentages)]

# Mover la leyenda más a la derecha
plt.legend(legend_labels, title="Años y Porcentajes", loc="center left", bbox_to_anchor=(1, 0.5))

# Mostrar el gráfico
plt.show()
```
